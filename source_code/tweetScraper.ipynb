```
list_of_tweets = []
users = ['NASA', 'NHL', 'MLB', 'spacegovuk', 'esa', 'NBA']
```

```
import tweepy
import csv
import pandas as pd
import time
import json
import datetime

#input your credentials here
access_token = "*******"  
access_token_secret = "*******"  
consumer_key = "*******"  
consumer_secret = "*******"

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth,wait_on_rate_limit=True)

user = users[5]
date_since = datetime.datetime(2020, 8, 31, 0, 0, 0)
actual_date = datetime.datetime(2020, 12, 4, 0, 0, 0)
max_id = None
while actual_date > date_since:
  max_count = 2500
  tweets = tweepy.Cursor(api.user_timeline,id=user, include_rts=False, max_id=max_id, tweet_mode='extended').items(max_count)
  while True:
      try:
          tweet = tweets.next()
          if tweet.in_reply_to_screen_name is not None:
            continue
          actual_date = tweet.created_at
          max_id = str(tweet.id - 1)
          if tweet.created_at < date_since:
            break
          print(user, tweet.full_text)
          list_of_tweets.append(tweet._json)
          max_count -= 1
          if max_count == 0:
            break
      except tweepy.TweepError:
          print(tweepy.TweepError)
          break
      except StopIteration:
          print('Sleeping')
          time.sleep(3)
```
